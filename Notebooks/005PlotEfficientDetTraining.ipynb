```python
import matplotlib.pyplot as plt
```

```python
# log_path = '../train1.log'
# start_epoch = 7
# log_path = '../train2.log'
# start_epoch = 1
# log_path = '../train3.log'
# start_epoch = 1
log_path = '../train4.log'
start_epoch = 1
```

```python
ap5095 = []
ap50 = []
ap75 = []
infile = open(log_path, 'r')
for line in infile:
    line = line.strip()
    parts = line.split(' = ')
    # print(line, parts[-1])
    if '(AP) @[ IoU=0.50:0.95 | area=   all | maxDets=100 ]' in line:
        ap5095.append(float(parts[-1]))
    elif '(AP) @[ IoU=0.50      | area=   all | maxDets=100 ]' in line:
        ap50.append(float(parts[-1]))
    elif '(AP) @[ IoU=0.75      | area=   all | maxDets=100 ]' in line:
        ap75.append(float(parts[-1]))
```

```python
x = range(start_epoch, start_epoch + len(ap5095), 1)
plt.plot(x, ap5095, label='AP IoU=0.50:0.95')
plt.plot(x, ap50, label='AP IoU=0.50')
plt.plot(x, ap75, label='AP IoU=0.70')
plt.legend()
plt.xlabel("Epochs")
plt.ylabel("Average Precision")
plt.title("Validation - AP")
plt.show()
```
