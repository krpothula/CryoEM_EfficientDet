```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
model_name = ['model4', 'model5', 'model6']
base_path = "/content/drive/MyDrive/UpWork/Protein/Yolo/Data/{}/results.csv"
data = []
for model in model_name:
  data_temp = pd.read_csv(base_path.format(model))
  data.append(data_temp)
data = pd.concat(data)
data['epochs'] = range(len(data))
```

```python
plt.plot(data['epochs'], data['     metrics/mAP_0.5'])
plt.legend()
plt.xlabel("Epochs")
plt.ylabel("Average Precision")
plt.title("Validation - AP")
plt.show()
```
